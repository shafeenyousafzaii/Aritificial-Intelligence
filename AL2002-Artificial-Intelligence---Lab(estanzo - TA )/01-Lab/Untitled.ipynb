## Task 4
- Generate a graph for 2x + 1 and cos x on the same plot comparing both of them

```python
X = np.linspace(0, 10, 10)
# Eq 1
Y = 2 * X + 1

# Eq 2
y = np.cos(X)

fig, ax = plt.subplots()

# Plot the first one
ax.plot(X, Y, '-o', label='2x + 1')

# Plot the second one
ax.plot(X, y, '-x', label='cos(x)')

# add a legend
ax.legend()
ax.set_title('Staright line and Cos')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

## Task 5
- Plot a Graph of 2 * x^2 and sin(x)

```python
X1 = np.linspace(0, 10, 10)
Y1 = 2 * (X1 ** 2)

Y2 = np.sin(X1)

fig, ax = plt.subplots()

ax.plot(X1, Y1, '-o', label='2x^2')
ax.plot(X1, Y2, '-x', label='sin(x)')

ax.legend()

ax.set_title('Graph of 2x^2 and sin(x)')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```
