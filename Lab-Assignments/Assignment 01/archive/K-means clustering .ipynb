```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("batting_bowling.csv")
df.head()
```

```python
df.info()
```

```python
df=df.drop(['PLAYER'],axis='columns')
df
```

```python
df=df.replace(to_replace='-',value='0')
df.head()
```

# performing the feature engineering

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df)
x=scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(x)
x=pca.transform(x)
```

# Using the elbow method determinig the number of clusters

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10,7))
plt.plot(range(1,11),wcss)
plt.show()
```

# By plotting the graph we came to know that there are 3 different types of clusters

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=0)
y_kmeans=kmeans.fit_predict(x)
```

# visualising the 3 different clusters

```python
plt.figure(figsize=(10,7))
plt.title('Clusters')
plt.xlabel('X1')
plt.ylabel('Y1')
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label='Cluster1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='yellow',label='Cluster2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='green',label='Cluster3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=400,c='magenta',label='Centroid')
plt.legend()
plt.show()
```
