```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

df=pd.read_csv("fruit_data_with_colors _1_.csv")
print(df)

data= df.drop(['fruit_name'], axis=1)
# print(data)


# data = df.fillna(df.mean())
mean_mass = data['mass'].mean()
data['mass'].fillna(mean_mass, inplace=True)


mean_width = data['width'].mean()
data['width'].fillna(mean_width, inplace=True)

mean_height = data['height'].mean()
data['height'].fillna(mean_height, inplace=True)




mean_colorScore = data['color_score'].mean()
data['color_score'].fillna(mean_colorScore, inplace=True)

# X=df[["Smoke","Lungs Capacity"]]
# y=df["Cancer"]


# Assuming the last column is the target variable
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)




k_values = range(1, 11)
accuracies = []


knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracies.append(accuracy)
# print(f"K = {k}, Accuracy = {accuracy}")


import matplotlib.pyplot as plt

plt.plot(k_values, accuracies, marker='o')
plt.title('Accuracy Score for Different K Values')
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()



```
